```python
from qiskit import *
from qiskit import QuantumCircuit
import numpy as np 
```

```python
def qpe(qc,n):

    
    for i in range(n-1):
        qc.h(i)
        angle = np.pi/2 * i 
        qc.crx(angle,i,n-1)


    return qc
```

```python
def inverse_qft(circuit, n):
    for i in range(n//2):
        circuit.swap(i, n-i-1)
    
    for i in reversed(range(n)):
        for j in reversed(range(i+1, n)):
            circuit.crx(-np.pi / 2**(j-i), j, i)
        circuit.h(i)
    
    return circuit
```

```python
n = 5
circuit = QuantumCircuit(n)
qc = qpe(circuit,n)
```

```python
qc.draw('mpl')
```

```python
qc2 = inverse_qft(qc,n-1)
qc2.draw('mpl')
```
